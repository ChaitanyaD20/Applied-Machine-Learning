5. Notice that some of the instances have a value of zero for the house price. Write two functions. One that computes the mean house price from the non-zero instances. One that computes the median house price for the non-zero instances. Note that you should not use the built-in median and mean functions to do this, but should compute it yourself. You can, however,compare the result of your functions to the built-in functions to verify your implementation.

```python
def mean_housing_price():
    mean=0.0
    total_sum=0.0
    for i in housing_data["price"]:
        total_sum=i+total_sum
    mean=total_sum/housing_data["price"].count()
    return mean    

mean_price=mean_housing_price()
print("Our calculated mean is:",mean_price)#Our function
print("In-built function value of mean is:",housing_data["price"].mean())#Inbuilt function
```

To calculate the mean of the price, we simply add up all values in price column and divide by the number of attributes.
To verify that the mean value is accurate, we have also checked with the in-built function of mean, and the values are accurate.

```python
def median_housing_price():
    
    median_housing_data=0.0
    price_list=[]
    for i in housing_data["price"]:
        price_list.append(i)
    price_list.sort()
    if len(price_list)%2==1:
        median_housing_data=price_list[len(price_list)//2+1]
    else:
        median_housing_data=(price_list[len(price_list)//2+2]+price_list[len(price_list)//2-1])/2
    return median_housing_data

median_price=median_housing_price()
print("Our calculated median is:",median_price)#Our function
print("In-built function value of median is:",housing_data["price"].median())#Inbuilt function
```

To calculate the median of the price, we sort all values in price column and then access the middle element of the sorted list. In the above case, our beginning index is 2(as our 1st filled row is number 2), and the python indices begin with 0, hence we need to adjust the value of 2.

8. Write your own function that computes the correlation coefficient between the target and the continuous valued features. Do the resulting correlation coefficients agree with your visual inspection of linear correlation? Why or why not? How do the correlation coefficients differ for each set (e.g. original, $0 priced homes removed, median replaced, and mean replaced)?

```python
def mean_calculator(data,variable):
    mean=0.0
    total_sum=0.0
    for i in data[variable]:
        total_sum=i+total_sum
    mean=total_sum/data[variable].count()
    return mean 


def summation(data,variable,price):
    sum=0.0
    mean=mean_calculator(data,variable)
    mean_price=mean_calculator(data,price)
    for i in range(len(data[variable])):
        sum=sum+((data[variable][i]-mean)*(data[price][i]-mean_price))
    return sum

    
def denominator(data,variable,price):
    sqrt_sum=0.0
    sum_x=0.0
    sum_y=0.0
    sum_xy=0.0
    mean=mean_calculator(data,variable)
    mean_price=mean_calculator(data,price)
    for i in range(len(data[variable])):
        sum_x=sum_x+((data[variable][i]-mean)**2)
        sum_y=sum_y+((data[price][i]-mean_price)**2)
    sqrt_sum=(sum_x*sum_y)**0.5
    
    return sqrt_sum

def correlation(data,variable,price):
    corr=0.0
    num=summation(data,variable,price)
    den=denominator(data,variable,price)
    corr=num/den
    return corr

print("For variable sqft_living, the 4 correlation values in order: 1. Original 2. Without Price 0$ 3. 0$ replaced with median and 4. 0$ replaced with mean are:")
print(correlation(housing_data,"sqft_living","price"))
print(correlation(housing_data_without_price_zero,"sqft_living","price"))
print(correlation(housing_data_with_median,"sqft_living","price"))
print(correlation(housing_data_with_mean,"sqft_living","price"))


print("For variable sqft_lot, the 4 correlation values in order: 1. Original 2. Without Price 0$ 3. 0$ replaced with median and 4. 0$ replaced with mean are:")
print(correlation(housing_data,"sqft_lot","price"))
print(correlation(housing_data_without_price_zero,"sqft_lot","price"))
print(correlation(housing_data_with_median,"sqft_lot","price"))
print(correlation(housing_data_with_mean,"sqft_lot","price"))


print("For variable sqft_above, the 4 correlation values in order: 1. Original 2. Without Price 0$ 3. 0$ replaced with median and 4. 0$ replaced with mean are:")
print(correlation(housing_data,"sqft_above","price"))
print(correlation(housing_data_without_price_zero,"sqft_above","price"))
print(correlation(housing_data_with_median,"sqft_above","price"))
print(correlation(housing_data_with_mean,"sqft_above","price"))

print("For variable sqft_basement, the 4 correlation values in order: 1. Original 2. Without Price 0$ 3. 0$ replaced with median and 4. 0$ replaced with mean are:")
print(correlation(housing_data,"sqft_basement","price"))
print(correlation(housing_data_without_price_zero,"sqft_basement","price"))
print(correlation(housing_data_with_median,"sqft_basement","price"))
print(correlation(housing_data_with_mean,"sqft_basement","price"))

```

11. For each data set (eight in total), us Scikit-learns KFold object to generate 10 folds of training and testing sets for cross validation. Be sure to set the random state variable. For each fold, train a Linear Regression model (see chapter 2 of ”Hands-on Machine Learning...”) using the training data. Then make predictions of the housing prices of the test data, using the trained model.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
def predict_regression(data):
    y=data[["price"]]
    X=data[["bedrooms","bathrooms","sqft_living","sqft_lot","floors","waterfront","view","condition","sqft_above","sqft_basement","yr_built","yr_renovated"]]
    kf = KFold(n_splits=10,random_state=42,shuffle=True)
    for train_index, test_index in kf.split(X):
        X_train= X.loc[train_index]
        X_test= X.loc[test_index]
        y_train= y.loc[train_index]
        y_test=y.loc[test_index]
    y_test=y_test.reset_index(drop=True)
    #print("Kfold Split Parameters are:",kf)
    lin_reg = LinearRegression()
    lin_reg.fit(X_train, y_train) 
    scores = cross_val_score(lin_reg,X,y,scoring="neg_mean_squared_error", cv=10)
    #print("\nCross Valuation Score is: ",list(scores))
    #print("Predictions on X_test are:\n\n",lin_reg.predict(X_test))
    #print("Actual Output data is:\n\n",y_test)
    housing_predictions = lin_reg.predict(X_test)
    lin_mse = mean_squared_error(y_test["price"], housing_predictions)
    return(lin_reg.predict(X_test),(y_test["price"]))
```

12. Write a function to compute the mean absolute error and mean square error for each fold of each data set. Report the results and discuss how different factors may have contributed to the final results.

```python
def mean_absolute_error(predicted_data,y_test_data):
    sum=0.0
    mae=0.0
    for i in range(len(y_test_data)):
        sum=sum+abs(predicted_data[i]-y_test_data[i])
    mae=sum/len(y_test_data)
    return mae[0]
```

```python
def mean_square_error(predicted_data,y_test_data):
    sum=0.0
    mse=0.0
    for i in range(len(y_test_data)):
        sum=sum+(predicted_data[i]-y_test_data[i])**2
    mse=sum/len(y_test_data)
    return mse[0]
```

Reference: https://www.statisticshowto.com/absolute-error/

The mean square error is calculated using the squared values of the difference between actual values and predicted values divided by the size of the test data set.

Reference: https://en.wikipedia.org/wiki/Mean_squared_error

2. In case of standardized values, we can have negative attributes as well, hence it is likely that the error values could be higher there as the range of the values is much greater than between 0 and 1.Hence the Mean Absolute and Mean Squared error values are higher.
3. In case of mean absolute error of normalized values, whenever we have 0 in the target variable of dataset, the error is slightly higher, because the difference between the predicted value and expected value becomes higher in this case. However when we remove the 0s or replace them with median or mean, the error value seems to be lesser, as it lessens the difference between predicted and expected value.
